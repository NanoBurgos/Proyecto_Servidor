# **🧭 Ruta recomendada**
- Linux básico
- SSH + UFW + Fail2Ban
- Networking
- Hardening
- Seguridad web
- Backups
- Monitoreo
- Seguridad avanzada

## **FASE 0 — Mentalidad (antes de tocar nada)**

Aprendé esto primero:

Seguridad ≠ instalar un antivirus

Seguridad = capas (defense in depth)

El mayor riesgo casi siempre es:

- malas configuraciones
- contraseñas débiles
- servicios expuestos innecesariamente

📌 Conceptos clave:

- Superficie de ataque
- Principio de mínimo privilegio
- Zero trust (no confiar en nada por defecto)

## **🟢 FASE 1 — Fundamentos de Linux (imprescindible)**

Antes de seguridad, dominá el sistema:

- Usuarios, grupos y permisos (chmod, chown)
- Estructura de Linux (/etc, /var, /home)
- Procesos (ps, top, htop)
- Servicios (systemctl)
- Networking básico (ip, netstat, ss)

📌 Objetivo:
-  Saber qué está corriendo en tu servidor en todo momento

## **🔐 FASE 2 — Seguridad básica del servidor (lo mínimo obligatorio)**

**1. Acceso seguro (SSH)**
- Desactivar login con contraseña
- Usar llaves SSH
- Cambiar puerto (opcional)
- Desactivar root login

Herramienta:
- OpenSSH (sshd_config)

**2. Firewall**

Configurar UFW

Ejemplo:

Permitir solo:
- SSH
- HTTP/HTTPS (si usás web)

📌 TODO lo demás: bloqueado

**3. Actualizaciones**

- Mantener sistema actualizado

Herramienta:

- unattended-upgrades

**4. Protección contra ataques automáticos**

- Instalar Fail2Ban

Bloquea:

- ataques de fuerza bruta SSH

## **🌐 FASE 3 — Networking y exposición (CRÍTICO)**

Aprendé:
- NAT
- Port forwarding
- IP pública vs privada
- DNS

📌 Clave:

👉 Si no está expuesto, no puede ser atacado

**Buenas prácticas:**

- No abrir puertos innecesarios
- Usar VPN en lugar de exponer servicios

Herramientas:

- WireGuard (recomendado)
- OpenVPN

## **🧱 FASE 4 — Hardening (endurecimiento del sistema)**

**1. Permisos y usuarios**
- No usar root
- Crear usuario con sudo limitado

**2. Servicios mínimos**
- Desinstalar todo lo que no uses

Comando clave:

            sudo ss -tulnp

**3. Seguridad del kernel**

Herramientas:

- AppArmor

Aprender:

- perfiles de seguridad
- sandboxing

**4. Protección de archivos**

Permisos estrictos en:
- /etc
- bases de datos
- backups

## **🗄️ FASE 5 — Seguridad de aplicaciones (MUY IMPORTANTE)**

Si vas a tener:

- web
- APIs
- base de datos

Debes aprender:

**1. Seguridad web (básico)**

- OWASP Top 10

Ejemplo:

- SQL Injection
- XSS
- CSRF

**2. Base de datos**
- No exponer DB a internet
- Usar contraseñas fuertes
- Limitar accesos por IP

**3. HTTPS obligatorio**
- Certificados SSL

Herramienta:

- Let’s Encrypt

## **🔍 FASE 6 — Monitoreo y detección**

**1. Logs**

Aprender a leer:

- /var/log/auth.log
- /var/log/syslog

**2. Detección de intrusos**

Herramientas:

- rkhunter
- chkrootkit

**3. Monitoreo en tiempo real**
- Netdata (opcional)
- Prometheus + Grafana (más avanzado)

## **💾 FASE 7 — Backups (LO MÁS IMPORTANTE)**

Si te hackean:

👉 lo único que te salva = backups

Estrategia:
- 3 copias:
     - local
     - externa
     - offline
 
Regla:
👉 “Si no está respaldado, ya está perdido”

## **🔒 FASE 8 — Seguridad avanzada**

**1. IDS/IPS**
- Snort
- Suricata

**2. Contenedores seguros**

Si usás Docker:

- Aislar contenedores
- No usar root
- Limitar recursos

**3. Reverse Proxy**
- NGINX

Ventajas:

- oculta servicios internos
- centraliza seguridad

## **🧪 FASE 9 — Pensar como hacker (nivel experto)**

Aprendé:

- Escaneo de puertos
- Enumeración
- Explotación básica

Herramientas:

- Nmap
- Wireshark

## **🧬 FASE 10 — Arquitectura segura (nivel pro)**

Diseñar sistema con:

- Segmentación de red
- DMZ (zona desmilitarizada)
- Separar:
    - servidor web
    - base de datos
    - backups

## **⚠️ ERRORES MÁS COMUNES (evitalos)**
- Usar contraseña en SSH ❌
- Abrir todos los puertos ❌
- No hacer backups ❌
- Exponer base de datos ❌
- No actualizar sistema ❌

## **💡 CONSEJO FINAL (esto vale oro)**

**Tu objetivo NO es:**

            “que nunca me hackeen”

**Tu objetivo real es:**

            👉 detectar rápido + minimizar daño + recuperar rápido
